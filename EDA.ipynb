```python
file_path = '/gdrive/MyDrive/DSCI 531/ArkansasJuv/08973-0001-Data.dta'

df = pd.read_stata(file_path)
```

```python
# 281: % white male, % white female, % black male, % black female (all juveniles)
#facilities where white male residents outnumber black male residents
exploratorydf = df[['V281', 'V282', 'V285', 'V286', 'V40']]
exploratorydfwb = exploratorydf[exploratorydf['V281']  > exploratorydf['V285']]
exploratorydfwb
```

```python
#facilities where black male residents outnumber white male residents
exploratorydf = df[['V281', 'V282', 'V285', 'V286', 'V40']]
exploratorydfbw = exploratorydf[exploratorydf['V281']  < exploratorydf['V285']]
exploratorydfbw
```

```python
import seaborn as sns
```

```python
securityb = sns.catplot(x="V285", y="V40", data = exploratorydfbw, kind="box")
securityb.fig.set_size_inches(9,6)
plt.title('Black Male Youth Population by Security Level for Majority Black Facilities')
plt.xlabel('Black Male Youth Population')
plt.ylabel('Facility Security Level')
plt.show()
```

```python
securityw = sns.catplot(x="V281", y="V40", data = exploratorydfwb, kind="box")
securityw.fig.set_size_inches(9,6)
plt.title('White Male Youth Population by Security Level for Majority White Facilities')
plt.xlabel('White Male Youth Population')
plt.ylabel('Facility Security Level')
plt.show()
```
